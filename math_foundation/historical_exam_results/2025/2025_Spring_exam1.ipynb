```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
```

```python
filename = 'sec1.csv'
grades = pd.read_csv (filename)
X = grades['Written Exam - 1 (2859867)'].to_numpy()
X = X[1:]
v1 = []
for x in X:
	if type(x).__name__ == 'str':
		v1.append(round(float(x)/500, 3))
```

```python
print(filename + ' mean score: %.3f'%np.mean(v1))
plt.hist(v1, bins=30, density=True, alpha=0.7)
plt.title(filename + ' Histogram')
plt.show()
```
